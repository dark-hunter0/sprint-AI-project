### Handling missing values

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
columns = ['ca', 'thal']
df_imputed = pd.DataFrame(imputer.fit_transform(X[columns]), columns=columns)

for column in X.columns:
    if column not in columns:
        df_imputed[column] = X[column]
    else:
        df_imputed[column] = df_imputed[column].astype('int64')
        
    
df_imputed['num'] = y
print(df_imputed.head())
df_imputed.to_csv("data/heart_disease.csv", index=False)
```

### Scaling the data 

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_columns= ['age', 'trestbps', 'chol', 'thalach','oldpeak']
df_imputed[scaled_columns] = scaler.fit_transform(df_imputed[scaled_columns])

df_imputed.head()
```
